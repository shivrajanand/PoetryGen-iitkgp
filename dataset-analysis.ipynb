```python
def cleaner(df):
    """Return rows with Hindi meaning and anuṣṭubh chandas only"""
    
    df = df[df["chanda"] == "anuṣṭubh"]

    text_cols = ["meaning"]
    mask = df[text_cols].astype(str).apply(
        lambda col: col.str.contains(r"[A-Za-z]", regex=True)
    ).any(axis=1)

    return df[~mask]

    
```

```python
train_clean = cleaner(train)
test_clean = cleaner(test)
val_clean = cleaner(val)
```

```python
train_clean.columns, test_clean.columns, val_clean.columns
```

```python
train_clean.shape, test_clean.shape, val_clean.shape
```

```python
train_clean.replace([r'^\s*$', '\xa0', '\u200b'], pd.NA, regex=True, inplace=True)
test_clean.replace([r'^\s*$', '\xa0', '\u200b'], pd.NA, regex=True, inplace=True)
val_clean.replace([r'^\s*$', '\xa0', '\u200b'], pd.NA, regex=True, inplace=True)


train_clean.dropna(axis=0, inplace=True)
test_clean.dropna(axis=0, inplace=True)
val_clean.dropna(axis=0, inplace=True)

print("Removed NULL ROWS")
```

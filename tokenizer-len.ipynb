```python
import pandas as pd
from transformers import AutoTokenizer
```

```python
train = pd.read_csv("C:/Users/shivr/OneDrive/Desktop/PoetryGen/dataset-anustubh-hindi/train.csv")
test = pd.read_csv("C:/Users/shivr/OneDrive/Desktop/PoetryGen/dataset-anustubh-hindi/test.csv")
val = pd.read_csv("C:/Users/shivr/OneDrive/Desktop/PoetryGen/dataset-anustubh-hindi/val.csv")
```

```python
MODEL_NAME = "microsoft/phi-4"
OUTPUT_PATH = "token-dist.txt"
```

```python
def token_dist(df_path):
    df = pd.read_csv(df_path)
    required_cols = {"prompt", "meaning"}
    if not required_cols.issubset(df.columns):
        raise ValueError(f"DataFrame must contain columns: {required_cols}")

    tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

    token_lengths = []

    for _, row in df.iterrows():
        text = str(row["prompt"]) + " " + str(row["meaning"])
        tokens = tokenizer.encode(text, add_special_tokens=False)
        token_lengths.append(len(tokens))
        
    token_series = pd.Series(token_lengths)

    stats = token_series.describe()

    with open(OUTPUT_PATH, "a") as f:
        f.write("=======================\n")
        f.write("\n\nToken Length Statistics\n")
        f.write(f"Data-File: {df_path}\n")
        for key, value in stats.items():
            f.write(f"{key}: {value}\n")
```

```python
token_dist("C:/Users/shivr/OneDrive/Desktop/PoetryGen/dataset-anustubh-hindi/train.csv")
```
